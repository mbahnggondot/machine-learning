```python
import pandas as pd

# Muat dataset
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

# Periksa nilai yang hilang pada 'day_df'
print("Nilai yang hilang pada 'day_df':")
print(day_df.isnull().sum())

# Periksa nilai yang hilang pada 'hour_df'
print("\nNilai yang hilang pada 'hour_df':")
print(hour_df.isnull().sum())

# Jika ada kolom yang tidak relevan pada kedua dataset, hapus kolom tersebut
# Misalnya, menghapus kolom yang tidak relevan (sesuaikan dengan kebutuhan)
# day_df = day_df.drop(columns=['unnecessary_column'])
# hour_df = hour_df.drop(columns=['unnecessary_column'])

# Isi nilai yang hilang atau hapus baris dengan nilai yang hilang pada kedua dataset
day_df = day_df.dropna()  # Menghapus baris dengan nilai hilang pada day_df
hour_df = hour_df.dropna()  # Menghapus baris dengan nilai hilang pada hour_df

# Periksa adanya duplikasi pada kedua dataset
print("\nDuplikasi pada 'day_df':", day_df.duplicated().sum())
print("Duplikasi pada 'hour_df':", hour_df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualisasi pola penggunaan sepeda berdasarkan waktu dalam satu hari
plt.figure(figsize=(12, 6))
sns.lineplot(data=hour_df, x='hr', y='cnt', marker='o')
plt.title('Pola Penggunaan Sepeda Berdasarkan Waktu dalam Satu Hari')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(range(0, 24))  # Mengatur label x untuk menunjukkan setiap jam
plt.grid(True)
plt.show()
```

```python
# Visualisasi pengaruh faktor cuaca terhadap jumlah penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.boxplot(data=day_df, x='weathersit', y='cnt')
plt.title('Pengaruh Faktor Cuaca terhadap Jumlah Penyewaan Sepeda')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Penyewaan')
plt.xticks([0, 1, 2, 3], ['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Memilih hanya kolom numerik dari day_df
numeric_cols = day_df.select_dtypes(include='number')

# Heatmap korelasi untuk kolom numerik dalam day_df
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriks Korelasi Fitur Numerik dalam day_df')
plt.show()
```

```python
# Histogram distribusi jumlah penyewaan sepeda
plt.figure(figsize=(8, 6))
sns.histplot(day_df['cnt'], bins=20, kde=True)
plt.title('Distribusi Jumlah Penyewaan Sepeda')
plt.xlabel('Jumlah Penyewaan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=day_df, x='temp', y='cnt')
plt.title('Hubungan antara Suhu dan Jumlah Penyewaan Sepeda')
plt.xlabel('Suhu (Normalized)')
plt.ylabel('Jumlah Penyewaan')
plt.show()
```

```python
# Menghitung rata-rata jumlah penyewaan per jam
average_hourly_usage = hour_df.groupby('hr')['cnt'].mean()

# Visualisasi pola penggunaan sepeda berdasarkan waktu dalam satu hari
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_hourly_usage.index, y=average_hourly_usage.values, marker='o')
plt.title('Rata-Rata Penggunaan Sepeda Berdasarkan Waktu dalam Satu Hari')
plt.xlabel('Jam')
plt.ylabel('Rata-Rata Jumlah Penyewaan')
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()

# Kesimpulan Pola Penggunaan
peak_hours = average_hourly_usage.idxmax()
print(f"Jam dengan penggunaan sepeda tertinggi: {peak_hours}:00")
```

```python
# Menghitung rata-rata jumlah penyewaan per kondisi cuaca
average_usage_weather = day_df.groupby('weathersit')['cnt'].mean()

# Visualisasi pengaruh faktor cuaca terhadap jumlah penyewaan sepeda
plt.figure(figsize=(10, 6))
sns.barplot(x=average_usage_weather.index, y=average_usage_weather.values)
plt.title('Rata-Rata Jumlah Penyewaan Sepeda Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-Rata Jumlah Penyewaan')
plt.xticks([0, 1, 2, 3], ['Clear', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'])
plt.show()

# Kesimpulan Pengaruh Cuaca
most_rentals_weather = average_usage_weather.idxmax()
print(f"Kondisi cuaca dengan jumlah penyewaan tertinggi: {most_rentals_weather}")
```
